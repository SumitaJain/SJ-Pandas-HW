```python
# Dependencies
import pandas as pd
```

```python
# load CSV
df = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv", low_memory=False)
df.head()
```

```python
# Inspect all columns
df.count()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
columns = [0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
# extract_df = df.iloc[columns]

# The hint above was not yielding the correct results
extract_df = df.iloc[:, columns]
extract_df.head()
```

```python
# extract_df.AttendedBootcamp.replace(to_replace=[0.0 ,1.0], value=['no', 'yes'])
# extract_df['AttendedBootcamp'] = extract_df['AttendedBootcamp'].apply(lambda x: 0.0 if x=='no' else 1)

# the hints above were not yielding the right results.
extract_df=extract_df.replace({0.0: "No", 1.0:"Yes"})
extract_df.head()
```

```python
# Calculate total number of respondents in survey
respondents = len(extract_df)
respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_bootcamp = extract_df.loc[extract_df["AttendedBootcamp"]=="Yes"]
attended_bootcamp.head()
```

```python
# Calculate average age of attendees
average_age = extract_df["Age"].mean()

# Calculate how many people attended a bootcamp
attendee_count = attended_bootcamp["AttendedBootcamp"].count()

# Calculate how many attendees hold degrees
degree_count = school_degrees = attended_bootcamp["SchoolDegree"].count()

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
gender_all = attended_bootcamp["Gender"].count()
male = attended_bootcamp["Gender"].value_counts()["male"]
female = attended_bootcamp["Gender"].value_counts()["female"]
non_binary = gender_all - male - female

# Calculate percentage of respondents who attended a bootcamp
attendee_percent = attendee_count / respondents * 100

# Calculate percentage of respondents belonging to each gender
# # I think these should be divided by attendee_count instead but the instructions (Comment above) asks for percent of respondents
male_percent = male / respondents * 100 
female_percent = female / respondents * 100
non_binary_percent = non_binary / respondents * 100

# Calculate percentage of attendees with a school degree
percent_degrees = degree_count / attendee_count * 100

# Calculate average post-bootcamp salary
average_post_salary = attended_bootcamp["BootcampPostSalary"].mean()
```

```python
# Create a new table consolodating above calculations
summary = [{"# Survey Respondents": respondents, "Average Age": average_age, "# Bootcamp Attendees": attendee_count, 
            "% Bootcamp Attendees": attendee_percent, "# Degrees": degree_count, "% Degrees": percent_degrees, 
            "# Male": male, "# Female": female, "# Non-Binary": non_binary, "% Male": male_percent, 
            "% Female": female_percent, "% Non-Binary": non_binary_percent, "Average Post-Bootcamp Salary in $": average_post_salary}]
summary_df = pd.DataFrame(summary)
summary_df = summary_df[["# Survey Respondents", "# Bootcamp Attendees", "% Bootcamp Attendees", "# Degrees", "% Degrees", "# Male", "# Female", "# Non-Binary", "% Male", "% Female", "% Non-Binary", "Average Post-Bootcamp Salary in $"]]
summary_df
```

```python
# Improve formatting before outputting spreadsheet
final_output = summary_df.round(2)
final_output
```
