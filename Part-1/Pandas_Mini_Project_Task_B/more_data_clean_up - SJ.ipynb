```python
# Dependencies
import pandas as pd
```

```python
# load CSV
# Read with pandas--low_memory required to suppress errors about mixed data types
df = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv", low_memory=False)
df.head(3)
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
columns = [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
extract_df = df.iloc[:, columns]
extract_df.head(3)
```

```python
# Change reading 0 and 1 to No and Yes, respectively
extract_df=extract_df.replace({0.0: "No", 1.0:"Yes"})
extract_df
```

```python
# Extract rows for only those who attended a bootcamp
bootcamp_attendees_df = extract_df.loc[(extract_df["AttendedBootcamp"] == "Yes")]
bootcamp_attendees_df.head(3)                     
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
bootcamp_name_groups_df = pd.DataFrame(bootcamp_attendees_df["BootcampName"].value_counts())
bootcamp_name_groups_df.reset_index(inplace=True)
bootcamp_name_groups_df.columns = ["Bootcamp Name", "# Attendees"]
bootcamp_name_groups_df
```

```python
## Based on results of the previous cell I selected 20 as the floor for 'significant number of attendees'
significant_bootcamps_df = bootcamp_name_groups_df.loc[(bootcamp_name_groups_df["# Attendees"] > 18)]
significant_bootcamps_df
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp

recommend_bootcamp = bootcamp_attendees_df.replace({"Yes": 1, "No": 0})
recommend_bootcamp = pd.DataFrame(recommend_bootcamp.groupby("BootcampName")["BootcampRecommend"].sum())

recommend_bootcamp.reset_index(inplace=True)
recommend_bootcamp.columns = ["Bootcamp Name", "# Recommenders"]

recommend_bootcamp.head()
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merged_df = pd.merge(bootcamp_name_groups_df, recommend_bootcamp, on="Bootcamp Name")
merged_df = merged_df.set_index("Bootcamp Name")
merged_df.head()
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
merged_df["% Recommenders"] = merged_df["# Recommenders"] / merged_df["# Attendees"] * 100

merged_df = merged_df.sort_values(["% Recommenders"], ascending=False)

## Tidying up the decimals (rounding to 2 places)
merged_df.round(2)
merged_df
```

```python
# Export to excel and remove index
merged_df.to_excel('output/pandas_mini_project_B_output.xlsx', index=False)
```
